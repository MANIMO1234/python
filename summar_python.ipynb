### 02 연산자
1) 산술연산자
   - 기본적으로 수학의 사칙연산
    ```
    a + b
    a - b
    a * b
    a / b
    ```
    ```
    a ** b  : a의 b제곱
    a // b  : a를 b로 나눈 몫
    a % b   : a를 b로 나눈 나머지 값
    ```

    ```
    divmod (a, b) :a를 b로 나눈 나머지와
                   몫을 함께 구해주는 함수이다
    ```
2) 비교연산자
   -  수학의 부등화와 같은 개념
  
   -  ```
      a > b
      a < b
      a >= b :a가 크거나 같다
      a <= b :a가 작거나 같다
      a == b :a 와 b의 값이 같다
      a =! b :a와 b의 값이 같지 않다
      ```
    - 부등호는 의문문으로 , 큰지 작은지 물어 보는 것이다

3) 논리연산자

   - not : 이건 아니야

### 04 시퀸스
- 단일한 데이터들이 묶여있는 단위
- 시퀀스는 데이터가 **순서대로** 나열된 자료구조
  (정렬 된 것과는 다르다)
  
1) list
   - 선언 : var_name =[val1, val2, val3]
   - 접근 : var_name[index]
     ex)
     ```
     location = ['서울', '대구', '부산']
     location[0] = '서울'
     location[1] = '대구'
     ```
2) tuple
   - 선언 : var_name = (value1,value2,value3)
   - 접근 : var_name[index]
   - 수정 불가능 하다(immutable) >> 계산결과를 불편하게 유지


3) range
    - range(n): 0 부터 n-1까지의 범위
    - range(n, m) : n부터 n-1까지의 범위 
    - range(n, m, s) : n부터 n-1까지 s만큼 증가하는 범위 
    * 만약 내가 10을 뽑고 십으면 , 11을 넣어야 함
    * 10번을 반복하고 싶으면 9를 넣으면 된다.
4) string
   - 그냥 문자

*인덱스 생략*
print(my_list[0:2]) -> 0번째 자료 부터 2번째 자료 까지
print(my_tuple[2:]) -> 2번째 자료에서 마지막까지
print(my_tuple[:3]) -> 3번째 자료 까지
print(my_range[-1]) -> 뒤에서 -1번째의 자료
print(my_string[::2]) ->데이터를 일련의 규칙성 있이 뽑음


   
 
### 05 시퀸스가 아닌 자료형

   - 중복값을 제가할때 유리하다

    ```
    print(my_set_a - my_set_b) # 차집합
    print(my_set_a | my_set_b) # 합집합
    print(my_set_a & my_set_b) # 교집합
    
    ```

    
2) Dictionary
    - 선언 : var_name = {key1: value1, key2: value2, key3: value3, .......}
    - 접근 : var_name[key]
      항상 key값을 중심으로 접근 할 수 있음

      -
      ```
      my_dict = {'서울': '02', '경기': '031'}
        print(my_dict['서울'])
        > 출력 값 02

      print(my_info['skill'][1])
      >>하나의 딕셔너리에 두개의 value값이 있는 경우
      또 한번의 순서를 부여 해즘
